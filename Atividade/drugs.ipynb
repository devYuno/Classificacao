```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = pd.read_csv("drug200.csv")
df
```

```python
df.isnull().sum()
```

```python
for i in df.columns:
    print(df[i].unique())
```

```python
features = ["Age", "Sex", "BP", "Cholesterol", "Na_to_K"]

target = df["Drug"]

features = df[features]
```

```python
for i in features:
    print(i)
```

```python
from sklearn.preprocessing import LabelEncoder

le_bp = LabelEncoder()
le_choles = LabelEncoder()

features['BP'] = le_bp.fit_transform(features['BP'])
features["Cholesterol"] = le_choles.fit_transform(features["Cholesterol"])
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

encoder_sex = ColumnTransformer(transformers=[('OneHot', OneHotEncoder(), [1])], remainder='passthrough')
features = encoder_sex.fit_transform(features)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(features, target, test_size = 0.3, random_state = 0)
```

```python
x_train.shape, x_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree

tree = DecisionTreeClassifier(criterion='gini', random_state=1)
tree.fit(x_train, y_train)

tree
```

```python
from sklearn.metrics import accuracy_score

previsao = tree.predict(x_test)
accuracy = accuracy_score(y_test, previsao)
```
